```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'  #    
#   
data = {
    'position': [i for i in range(-10, 11)],  # x  (-10 ~ 10)
    'unobserved': [5, 8, 15, 25, 35, 40, 45, 50, 45, 35, 25, 15, 10, 8, 6, 5, 4, 3, 2, 1, 1],
    'observed':   [3, 4, 5, 6, 7, 8, 10, 12, 12, 12, 12, 12, 12, 12, 10, 8, 7, 6, 5, 4, 3]
}

df = pd.DataFrame(data)

# 
plt.figure(figsize=(10, 5))
plt.plot(df['position'], df['unobserved'], label=' ()', marker='o')
plt.plot(df['position'], df['observed'], label=' ()', marker='x')
plt.xlabel(' (position)')
plt.ylabel(' ')
plt.title('  ')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
